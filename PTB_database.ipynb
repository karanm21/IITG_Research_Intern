```python
from IPython.display import display
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import os
import shutil
import posixpath
import pandas as pd
import wfdb
```

```python
wfdb.dl_database('ptbdb', os.path.join(os.getcwd(), 'ptbdb'))
```

```python
record = wfdb.rdrecord('ptbdb/patient001/s0010_re')
wfdb.plot_wfdb(record=record, title='Patient 1')
display(record.__dict__)
```

```python
signals, fields = wfdb.rdsamp('ptbdb/patient001/s0010_re', sampfrom=100, sampto=15000)
display(signals)
display(fields)
```

```python
record = wfdb.rdrecord('ptbdb/patient001/s0010_re', sampto = 15000)
annotation = wfdb.rdann('ptbdb/patient001/s0010_re', 'dat', sampto = 15000)

wfdb.plot_wfdb(record=record, annotation=annotation, title='ptbdb database', time_units='seconds')
```
