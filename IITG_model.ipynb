```python
from keras.models import Sequential
from keras.layers import Conv1D, MaxPooling1D, Flatten, Dense
```

1D CNN MODEL

```python
num_classes=3
```

```python
model_1D= Sequential()
model_1D.add(Conv1D(32, kernel_size=3, activation='relu', input_shape=(260, 1)))
model_1D.add(MaxPooling1D(pool_size=2))
model_1D.add(Conv1D(64, kernel_size=3, activation='relu'))
model_1D.add(MaxPooling1D(pool_size=2))
model_1D.add(Flatten())
model_1D.add(Dense(250, activation='relu'))
model_1D.add(Dense(100, activation='relu'))
model_1D.add(Dense(50, activation='relu'))
model_1D.add(Dense(num_classes, activation='softmax'))
```

```python
model_1D.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model_1D.summary()
```

```python
model_1D.fit(X_train, y_train, batch_size=128, epochs=100, validation_data=(X_test, y_test))
```

DNN MODEL

```python
model = keras.Sequential()
model.add(keras.Input(shape=(260,)))
model.add(layers.Dense(300,activation='relu'))
model.add(layers.Dense(100,activation='relu'))
model.add(layers.Dense(50,activation='relu'))
model.add(layers.Dense(20,activation='relu'))
model.add(layers.Dense(3,activation='softmax'))

opt = optimizers.Adam(0.0001)
model.compile(optimizer=opt, loss=losses.categorical_crossentropy, metrics=['accuracy'])

model.summary()
```

```python
model.fit(X_train, y_train,batch_size = 128,epochs = 100, validation_data=(X_test, y_test))
```

Accuracy of dnn on 260 beats size

```python
model.evaluate(X_test, y_test)
```

```python
from sklearn.metrics import multilabel_confusion_matrix
```

```python
pred_dnn=model.predict(X_test)
pred_dnn = np.round(pred_dnn)
pred_dnn
```

```python
cms = multilabel_confusion_matrix(y_test, pred_dnn)
cms
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
class_labels = ['Class 1', 'Class 2', 'Class 3']
```

```python
# Plot confusion matrices
num_classes = len(class_labels)
fig, axes = plt.subplots(num_classes, 1, figsize=(6, 6*num_classes))

for i, (cm, class_label) in enumerate(zip(cms, class_labels)):
    ax = axes[i]
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False, ax=ax)
    ax.set_xlabel('Predicted')
    ax.set_ylabel('True')
    ax.set_title(f'Confusion Matrix for {class_label}')

plt.tight_layout()
plt.show()
```
